```python
# sample code to scrap data from any website using beautiful soup we can implement
# same algo in scraping data from website to collect information on COVID-19
from bs4 import BeautifulSoup as bs
import requests
```

```python
link = "https://www.amazon.in/Apple-MacBook-Chip-13-inch-256GB/product-reviews/B08N5VSQNG/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews"
```

```python
page = requests.get(link)
```

```python
page
```

```python
page.content
```

```python
soup = bs(page.content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
user_names = soup.find_all('span', class_='a-profile-name') # list of all users
```

```python
user_names
```

```python
customer_name = []
for i in range(0, len(user_names)):
    customer_name.append(user_names[i].get_text())
customer_name
```

```python
review_title = soup.find_all('a',class_='review-title-content')
review_title
```

```python
title = []
for i in range (0, len(review_title)):
    title.append(review_title[i].get_text())
title
```

```python
title[:] = [k.lstrip('\n') for k in title]
title[:] = [k.rstrip('\n') for k in title]
title
```

```python
html_name = soup.find_all('span', class_='a-profile-name')
html_name.pop(0)
html_name.pop(0)
names = []
for i in range (0,len(html_name)):
    names.append(html_name[i].get_text())
# len(names)
```

```python
html_review_votes = soup.find_all('span', {"data-hook":'helpful-vote-statement'})
votes = []
for i in range (0,len(html_review_votes)):
    votes.append(html_review_votes[i].get_text())
# votes
```

```python
html_review_date = soup.find_all('span', {'data-hook':'review-date'})
# html_review_date
review_date = []
for i in range (0,len(html_review_date)):
    review_date.append(html_review_date[i].get_text())
# review_date
```

```python
html_review_content = soup.find_all('span', {'data-hook': 'review-body'})
review_content = []
for i in range(0, len(html_review_content)):
    review_content.append(html_review_content[i].get_text())
# review_content
```

```python
review_content = [k.lstrip('\n\n  ') for k in review_content]
review_content = [k.rstrip('\n\n') for k in review_content]
# review_content[0]
```
